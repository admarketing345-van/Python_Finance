```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
ticker = "SPY"
start_date = '2020-01-01'
end_date = '2025-10-01'
df = yf.download(ticker, start_date, end_date, auto_adjust=True)
df.columns = df.columns.droplevel(1)
```

```python
df['return'] = df['Close'].pct_change()
df['cum return'] = (1 + df['return']).cumprod() - 1 
df = df.dropna()
df.head(5)
```

```python
#vissualize the return
plt.figure(figsize=(10,6))
plt.plot(df['cum return'])
plt.title(f'{ticker} return')
plt.xlabel('Date')
plt.ylabel('return')
plt.grid(True)
plt.show
```
